```python
df.to_sql("SPACEXTBL", conn, if_exists='replace', index=False, method="multi")
```

```python
cursor = conn.cursor()
```

```python
cursor.execute("create table SPACEXTABLE as select * from SPACEXTBL where Date is not null")
```

```python
cursor.execute("select* from SPACEXTBL")
results = cursor.fetchall()
for row in results:
    print(row)
```

### Task 1




##### Display the names of the unique launch sites  in the space mission


```python
cursor.execute("SELECT DISTINCT `Launch site` FROM SPACEXTBL")
results = cursor.fetchall()
for row in results:
    print(row)
```


### Task 2


#####  Display 5 records where launch sites begin with the string 'CCA'


```python
cursor.execute('SELECT * FROM SPACEXTBL WHERE "Launch site" LIKE "CCA%" LIMIT 5')
results1 = cursor.fetchall()
for row in results1:
    print(row)
```

### Task 3




##### Display the total payload mass carried by boosters launched by NASA (CRS)


```python
cursor.execute('SELECT "Payload mass" FROM SPACEXTBL ')
results1 = cursor.fetchall()
for row in results1:
    print(row)
```

### Task 4




##### Display average payload mass carried by booster version F9 v1.1


```python
cursor.execute('SELECT AVG("Payload mass") FROM SPACEXTBL WHERE "Version Booster" = \'F9 v1.1\' ')
results1 = cursor.fetchall()
for row in results1:
    print(row)
```

### Task 5

##### List the date when the first succesful landing outcome in ground pad was acheived.



```python
cursor.execute('SELECT MIN("Date") FROM SPACEXTBL WHERE "Booster landing" = \'1\'')
results1 = cursor.fetchall()
for row in results1:
  print(row)
```
