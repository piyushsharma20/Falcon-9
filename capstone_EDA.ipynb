```python
landing_outcomes = df["Booster landing"].value_counts()
landing_outcomes
```

```python
for i,outcome in enumerate (landing_outcomes.keys()):
  print(i,outcome)
```

```python
bad_outcomes=set(landing_outcomes.keys()[[1,2,4,5,6,7]])
bad_outcomes
```

```python
landing_class = []
for key, value in df['Booster landing'].items():
     if value in bad_outcomes:
        landing_class.append(0)
     else:
        landing_class.append(1)
```

```python
df["Booster landing"] = landing_class
```
